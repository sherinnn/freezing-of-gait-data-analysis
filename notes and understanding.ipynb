The online device is a wearable, context-aware assistive system designed to detect freezing of gait (FOG) in real time (i.e., online, while the patient is walking) and immediately provide auditory cues to help Parkinson’s disease patients resume walking.
“Online” here means:
Detection and decision-making happen in real time
No post-processing or offline analysis is required
Feedback is delivered during the actual freezing episode

The device continuously computes:

Freeze Index (FI)= Power in 0.5–3 Hz band / Power in 3–8 Hz band

High FI → likely FOG
Low FI → normal walking
